```python
import numpy as np
```

```python
fn = "day9.txt"

data = open(fn).read().strip()
sequences = [np.asarray(x.split(), dtype=np.int64) for x in data.split('\n')]

def diff_result(x,y):
    return(x+y)

def diff_result_2(x,y):
    return(x-y)

result = []
for seq in sequences:
    print('='*50)
    diff_hold = []
    diff_hold.append(seq)
    while np.any(diff_hold[-1] != 0):
        diff_hold.append(np.diff(diff_hold[-1]))
    #Reverse values in 
    for i, val in enumerate(diff_hold):
        diff_hold[i] = np.flip(val) 
    # diff_hold.append(seq)
    iter_result = [0]
    print(diff_hold)
    for i, diff_arr in enumerate(np.flip(diff_hold)):
        # print('-'*50)
        # print(diff_arr)
        # iter_result.append(diff_result_2(iter_result[-1], diff_arr[-1])) 
        iter_result.append(diff_result_2(diff_arr[-1],iter_result[-1])) 
        print(iter_result)
    result.append(iter_result[-1]) 
print(np.sum(result))
```
