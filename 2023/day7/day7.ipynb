```python
fn = 'day7.txt'
data = open(fn).read().strip()

hands = [x.split(' ')[0] for x in data.split('\n') ]
bets = [int(x.split(' ')[1]) for x in data.split('\n')]
card_val = {'A': 14, 'K': 13, 'Q': 12, 'J': 0, 'T': 10, '9':9,
            '8':8, '7':7, '6':6, '5':5, '4':4, '3':3, '2':2}

hands_1 = dict()
bets_1 = dict() 
#Sett up hand strengths
for i in np.arange(7):
    hands_1[i] = []
    bets_1[i] = []
for i, hand in enumerate(hands):
    chk = 0 #check hand before and after

    #Use number of card counts instead
    num_card_val, cnts = np.unique(list(hand),return_counts=True)
    max_card_val = num_card_val[np.argmax(cnts)]
    chk= 0
    if max_card_val == 'J':
        if len(cnts) == 1: #5 of J 
            pass
        else:
            num_card_val = np.delete(num_card_val, (np.argmax(cnts)))
            cnts = np.delete(cnts, (np.argmax(cnts)))
            max_card_val = num_card_val[np.argmax(cnts)]
            chk=1
   
    #Hand to use for hand strength 
    temp_hand = hand.replace('J', str(max_card_val)) #part2 


    #Hand to use for sorting
    hand = hand.replace('T', chr(ord('9')+1))
    hand = hand.replace('J', chr(ord('1')-2)) #part 2
    hand = hand.replace('Q', chr(ord('9')+3))
    hand = hand.replace('K', chr(ord('9')+4))
    hand = hand.replace('A', chr(ord('9')+5))

    #cnt num of cards for hand strength
    cards, crd_cnt = np.unique(list(temp_hand), return_counts=True)

    if len(crd_cnt) == 1: #five of a kind
        hands_1[6].append((hand,bets[i]))
    elif len(crd_cnt) == 2 and np.max(crd_cnt)==4: #four of a kind
        hands_1[5].append((hand,bets[i]))
    elif len(crd_cnt) == 2 and np.max(crd_cnt)==3: #fullhouse
        hands_1[4].append((hand,bets[i]))
    elif (len(crd_cnt) == 3) and np.max(crd_cnt)==3: #three of a kind
        hands_1[3].append((hand,bets[i]))
    elif (len(crd_cnt) == 3) and np.max(crd_cnt)==2: #two pair
        hands_1[2].append((hand,bets[i]))
    elif (len(crd_cnt) == 4): #one pair
        hands_1[1].append((hand,bets[i]))
    else: #one pair
        hands_1[0].append((hand,bets[i]))

#Sort cards for each hand strength key
chk = 1
ans = 0 
for key in hands_1:
    inter = sorted(hands_1[key])
    for k in inter:
        ans += chk*k[1] 
        chk +=1

print(ans)
   
    
#
```
