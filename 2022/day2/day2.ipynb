```python
rps_map = {'A': 1, 'B':2, 'C':3, 'X': 1, 'Y':2, 'Z':3}

fn = 'day2.txt'

D = open(fn).read().strip()

ans = 0
for line in D.split('\n'):
    opp_play = rps_map[line[0]]
    my_play = rps_map[line[2]]
    # print("="*8)
    # print(opp_play)
    # print(my_play)
    if opp_play == my_play: #draw
        ans += my_play
        ans += 3

    #Winning throws
    elif (my_play == 1) & (opp_play == 3 ):
        ans += my_play
        ans += 6 

    elif (my_play == 2) & (opp_play == 1 ):
        ans += my_play
        ans += 6 

    elif (my_play == 3) & (opp_play == 2 ):
        ans += my_play
        ans += 6 

    else:
        ans += my_play

print(ans)
       
        
```

```python
#Part two
rps_map = {'A': 1, 'B':2, 'C':3, 'X': 1, 'Y':2, 'Z':3}

fn = 'day2.txt'

D = open(fn).read().strip()

ans = 0
for line in D.split('\n'):
    opp_play = rps_map[line[0]]
    my_play = rps_map[line[2]]
    # print("="*8)
    # print(opp_play)
    # print(my_play)
    if my_play == 1: #Lose
        print('0')
        if opp_play == 1: 
            ans += 3
        if opp_play == 2: 
            ans += 1 
        if opp_play == 3: 
            ans += 2 
    if my_play == 2: # Draw
        ans += 3
        ans += opp_play
    if my_play == 3: #Win 
        print('2')
        if opp_play == 1: 
            ans += 2 
        if opp_play == 2: 
            ans += 3 
        if opp_play == 3: 
            ans += 1 
        ans += 6
print(ans)
       
        
```
